```python
# Import the library
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import  confusion_matrix, accuracy_score, classification_report
```

```python
df = pd.read_csv('/content/Machine-Learning-Project/Advertising_data.csv')
X = df.iloc[:,[2,3]].values
y = df.iloc[:,4].values
```

```python
df.head()
```

```python
# Splitting the dataset into training and test data
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=5)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Fitting the LogisticRegression model on training data
clf = LogisticRegression(penalty='l2', random_state= 5, solver='lbfgs', max_iter=10, multi_class='deprecated')
clf.fit(X_train,y_train)
```

```python
# Predict the results of test set
y_pred = clf.predict(X_test)
#y_pred
```

```python
# Making the confusion matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

## Define the Hyperparameters grid for tunning

```python
parameters = {
    'C': np.logspace(-4,4,20),
    'penalty': ['l1','l2','elasticnet','None'],
    'solver':['newton-cg','lbfgs','sag','saga'],
    'max_iter':[100,500,1000,1500] }
parameters
```

## Apply GridSearchCV to find the beat hyperparameters

```python
grid_search = GridSearchCV(clf,param_grid=parameters,cv=10,verbose=1,n_jobs=-1)
grid_search.fit(X_train,y_train)
```

```python
accuracy = grid_search.best_score_

accuracy
```

```python
## Display the best parameter values

grid_search.best_params_
```

```python
# Fitting again the LogisticRegression model on training data
clf = LogisticRegression(penalty='l1', solver='saga', max_iter=100)
clf.fit(X_train,y_train)
```

```python
# Predict the test set results
y_pred = clf.predict(X_test)
```

```python
# Making the confusion matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
accuracy=accuracy_score(y_test,y_pred)
accuracy
```
